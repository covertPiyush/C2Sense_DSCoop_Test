### Importing all the required libraries for this assignment

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from statsmodels.stats import weightstats as stests
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
```

### A class which defines a function to plot the count of orders week wise

```python
class Plotter():

    def  plot_count_orders_per_week(self,orders):
        '''
        params : A dataframe which basically represents the entire CSV file
        returns : A plot which is separated by the week number inside every month. Read Jan -1 as 1st week of Jan and so forth.
        '''
        month_replace = []
        month_map = {
            1: 'Jan',
            2: 'Feb',
            3: 'Mar',
            4: 'Apr',
            5: 'May',
            6: 'Jun',
            7: 'Jul',
            8: 'Aug',
            9: 'Sep',
            10: 'Oct',
            11: 'Nov',
            12: 'Dec'
        }
        orders['date'] = pd.to_datetime(orders['date'])
        orders['weekNum'] = pd.to_datetime(orders['date']).apply(lambda d: (d.day - 1) // 7 + 1)
        orders['month'] = orders['date'].apply(lambda x: x.date().month)
        for elem in orders['month']:
            for k, v in month_map.items():
                if elem == k:
                    elem = v
                    month_replace.append(elem)
        orders['month'] = month_replace
        orders['weekNum'] = orders['weekNum'].astype(str)
        orders['weekNum'] = orders['month'] + '-' + orders['weekNum']
        r = orders.groupby('weekNum', sort=False).count().reset_index()

        figure(num=None, figsize=(40, 20), dpi=80, facecolor='w', edgecolor='k')
        x = r['weekNum'].values.tolist()
        y = r['value'].values.tolist()
        plt.plot(x, y)
        plt.xticks(rotation=70, fontsize=20)
        plt.yticks(fontsize=20)
        plt.show()

        return r
```

```python
def read_csv():
    '''
    params: No input
    returns : A dataframe which represents the uploaded file
    '''
    orders = pd.read_csv('screening_exercise_orders_v201810.csv')
    return orders
```

```python
def part_a(orders):
    '''
    params : A dataframe which represents the uploaded file
    returns : The output for Task A which is an assembled dataframe containing the number of orders placed by each customer and the
              last placed order date in an ascending order sorted by customer_id.    
    '''
    cust_list = list(set(orders['customer_id'].values.tolist()))
    data_list = []
    for elem in cust_list:
        filtered_rows = orders[orders['customer_id'] == elem]
        gender = filtered_rows['gender'].iloc[0]
        most_recent_date = filtered_rows['date'].max()
        order_count = len(filtered_rows)
        data_list.append([{'customer_id': elem, 'gender': gender, 'most_recent_order_date': most_recent_date,'order_count': order_count}])
    assembled_dataframe = pd.DataFrame()
    for data in data_list:
        if len(assembled_dataframe) == 0:
            assembled_dataframe = pd.DataFrame(data)
        else:
            assembled_dataframe = assembled_dataframe.append(pd.DataFrame(data))
    return assembled_dataframe.sort_values('customer_id')
```

```python
def part_b(orders):
    '''
    params : A dataframe which represents the uploaded file
    returns : A plot showing the order count per week. The format is Jan-1 which means 1st week of January.
    '''
    r = Plotter().plot_count_orders_per_week(orders)
    return r
```

```python
def part_c(orders):
    '''
    params : A dataframe which represents the uploaded file
    returns :  The mean difference between the order value of both the genders(0,1), the test statistic for checking the statistical
               significance of the difference between the average order value of both the gender, the probability value which determines 
               whether to accept or reject the null hypothesis.
    '''
    gender0 = orders.loc[orders['gender'] == 0]
    gender1 = orders.loc[orders['gender'] == 1]
    mean_order_value_gender0 = gender0['value'].sum()/len(gender0)
    mean_order_value_gender1 = gender1['value'].sum()/len(gender1)
    mean_difference = mean_order_value_gender0 - mean_order_value_gender1
    ztest, pval = stests.ztest(gender0['value'],gender1['value'],value=0,alternative='two-sided')
    return mean_difference,ztest,pval
```

```python
def part_d(orders):
    '''
    params : A dataframe which represents the uploaded file
    returns : A confusion matrix, a ROC-AUC curve determining the quality of the predictions
    '''
    gender = orders[['gender','predicted_gender']]
    true_positive = 0
    false_negative = 0
    true_negative = 0
    false_positive = 0
    confusion_matrix = np.matrix([[0, 0], [0, 0]])
    for i, row in gender.iterrows():
        if row['gender'] == 0 and row['predicted_gender'] == 0:
            true_positive += 1
        elif row['gender'] == 0 and row['predicted_gender'] == 1:
            false_negative += 1
        elif row['gender'] == 1 and row['predicted_gender'] == 1:
            true_negative += 1
        elif row['gender'] == 1 and row['predicted_gender'] == 0:
            false_positive += 1
    confusion_matrix[0, 0] = true_positive
    confusion_matrix[0, 1] = false_negative
    confusion_matrix[1, 0] = false_positive
    confusion_matrix[1, 1] = true_negative
    false_positive_rate,true_positive_rate,threshold = roc_curve(gender['gender'],gender['predicted_gender'])
    roc_auc = auc(false_positive_rate,true_positive_rate)
    plt.title('Receiver Operating Characteristic')
    plt.plot(false_positive_rate, true_positive_rate, 'b',
    label='AUC = %0.2f'% roc_auc)
    plt.legend(loc='lower right')
    plt.plot([0,1],[0,1],'r--')
    plt.xlim([-0.1,1.2])
    plt.ylim([-0.1,1.2])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    return confusion_matrix,false_positive_rate,true_positive_rate,threshold,roc_auc
```

```python
print('Difference in mean order value for gender 0 and gender 1 is %.2f' %(mean_difference))
print('z-test-statistic for checking the statistical significance of the difference in mean is %.2f' %(z_test_statistic))
print('The probability value to accept or reject the null hypothesis is %2f percent' %(p_val*100))
```

Let us write a hypothesis for checking the statistical significance of how different the means of average order value of both the genders(0,1) are. 

#### H0 : The difference between mean average order value of both the groups(0,1) is not statistically significant

#### H1 : The difference between mean average order value of both the groups(0,1) is statistically significant

The distribution of the both the groups is right skewed. Z-test factors in for skewed distribution(not normally distributed) if the sample size is large. A p-value of 4.8% helps us to reject the null hypothesis which states that the difference is statistically significant between the mean order value of gender 0 and gender 1.

Assumptions:
The population size is not that large for each group(namely 6759 for gender 0, 6712 for gender 1). Assuming that I want to generalize to the entire world, I would take my dataset as the sample. Here, the population variance is known. Using a sample size greater than 30 when the population variance is known, I have proceeded to do a z-test with 95% confidence interval and 5% of statistical significance. A p-value less than 5% would help us to reject the null hypothesis.
